```python
import ast
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
import json

data = pd.DataFrame(columns=['metric','threshold','latency','server','s_acc','ms_acc','m_acc'])

fig = plt.figure(figsize=(20,10))
axs = fig.subplots(nrows=2, ncols=4)

for i in range(len(os.listdir('../ Kaggle/Final'))):

    filename = os.listdir('../ Kaggle/Final')[i]
    file = open('../ Kaggle/Final/'+filename, 'r')
    lines = file.readlines()
    file.close()

    for m in range(4):  #metric

        latency = np.zeros(6)
        s = np.zeros(6)
        s_acc = np.zeros(6)
        ms_acc = np.zeros(6)
        m_acc = np.zeros(6)

        for d in range(3): #datasets

            temp = json.loads(lines[1+(1*d)+(4*d)+m])
            latency += np.array(temp['latency'])
            s += np.array(temp['server'])
            s_acc += np.array(temp['s_acc'])
            ms_acc += np.array(temp['ms_acc'])
            m_acc += np.array(temp['m_acc'])

        #metric.append(temp['metric'])   
        threshold = temp['threshold']
        latency = latency/3
        s = s/3
        acc = (m_acc +s_acc)/(10000*3)
        axs[i//4,i%4].plot(threshold,acc,marker='o',label = temp['metric'])
        axs[i//4,i%4].legend()

    axs[i//4,i%4].set_xlabel('threshold')
    axs[i//4,i%4].set_ylabel('Total Accuracy')
    axs[i//4,i%4].set_title('Metrics Comparison for '+filename[:-4])

plt.show()
```
